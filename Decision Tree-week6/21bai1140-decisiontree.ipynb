```python
train_data = pd.read_csv("/kaggle/input/digit-identification-with-decision-tree-b1/train.csv",header=None)
train_data
```

```python
test_data = pd.read_csv("/kaggle/input/digit-identification-with-decision-tree-b1/test.csv",header=None)
test_df = test_data
```

```python
#Dropping the first columns
train_data=train_data.drop(columns=[0])
test_data=test_data.drop(columns=[0])
```

```python
#Drop the last target column
X_train=train_data.drop(columns=[65])
y_train=train_data[65]
```

```python
# Split the training data into training and validation sets
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.3, random_state=42)
```

```python
# Preprocess the data: Scale pixel values to a range between 0 and 1
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Train the Decision Tree classifier with different hyperparameters
model = DecisionTreeClassifier(max_depth=10, criterion='gini', min_samples_split=2)
model.fit(X_train, y_train)
```

```python
# Predict the labels for the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the accuracy of the model on the validation set
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on validation set: {accuracy}")
```

```python
# Train the Decision Tree classifier with different hyperparameters
model3 = DecisionTreeClassifier(max_depth=8, criterion='gini', min_samples_split=2, min_samples_leaf=1, max_leaf_nodes=None, random_state=42)
model3.fit(X_train, y_train)
# Predict the labels for the test set
y_pred = model3.predict(X_test)
# Evaluate the accuracy of the model on the validation set
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on validation set: {accuracy}")
```

```python
# Train the Decision Tree classifier with different hyperparameters
model4 = DecisionTreeClassifier(max_depth=10, criterion='gini', min_samples_split=2, min_samples_leaf=1, max_leaf_nodes=None,max_features='sqrt')
model4.fit(X_train, y_train)
# Predict the labels for the test set
y_pred = model4.predict(X_test)
# Evaluate the accuracy of the model on the validation set
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on validation set: {accuracy}")
```

```python
# Train the Decision Tree classifier with different hyperparameters
model5 = DecisionTreeClassifier(max_depth=10, criterion='gini', min_samples_split=2, min_samples_leaf=1, max_leaf_nodes=None,max_features='log2')
model5.fit(X_train, y_train)
# Predict the labels for the test set
y_pred = model5.predict(X_test)
# Evaluate the accuracy of the model on the validation set
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on validation set: {accuracy}")
```

```python
# Train the Decision Tree classifier with different hyperparameters
model6 = DecisionTreeClassifier(max_depth=10, criterion='gini', min_samples_split=2, min_samples_leaf=1, max_leaf_nodes=None,max_features=10)
model6.fit(X_train, y_train)
# Predict the labels for the test set
y_pred = model6.predict(X_test)
# Evaluate the accuracy of the model on the validation set
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on validation set: {accuracy}")
```

```python
# Train the Decision Tree classifier with different hyperparameters
model2 = DecisionTreeClassifier(max_depth=10, criterion='gini', min_samples_split=2, min_samples_leaf=1, max_leaf_nodes=None)
model2.fit(X_train, y_train)
# Predict the labels for the test set
y_pred = model2.predict(X_test)
# Evaluate the accuracy of the model on the validation set
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on validation set: {accuracy}")
```

```python
# Once you're satisfied with the model's performance, you can make predictions on the test data
X_test = scaler.transform(test_data)  # Preprocess the test data using the same scaler
y_pred_test = model2.predict(X_test)
```

```python
#Create the submission file
submission = pd.DataFrame({"ID": test_df[0], "Category": y_pred_test})
submission.to_csv("submission.csv", index=False)
```
